```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Carregar os dados dos arquivos CSV
df_dolar = pd.read_csv('../dolar/dolar-2023-03-21.csv')
df_euro = pd.read_csv('../euro.csv')

# Converter a coluna de data para o tipo datetime e definir como índice
df_dolar['Data'] = pd.to_datetime(df_dolar['Data'])
df_euro['Data'] = pd.to_datetime(df_euro['Data'])
df_dolar.set_index('Data', inplace=True)
df_euro.set_index('Data', inplace=True)

# Certifique-se de que os dados estão em uma frequência que a seasonal_decompose possa usar (por exemplo, diária, mensal)
# Se necessário, reamostrar os dados para a frequência desejada
df_dolar = df_dolar.resample('D').mean()
df_euro = df_euro.resample('D').mean()
```

```python
# Aplicar seasonal_decompose
resultados_dolar_additive = seasonal_decompose(df_dolar['Valor'], model='additive')
resultados_euro_additive = seasonal_decompose(df_euro['Valor'], model='additive')
```

```python
# Aplicar seasonal_decompose
resultados_dolar_multiple = seasonal_decompose(df_dolar['Valor'], model='multiple')
resultados_euro_multiple = seasonal_decompose(df_euro['Valor'], model='multiple')
```
