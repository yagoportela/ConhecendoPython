```python
import pandas as pd
from scipy.stats import zscore
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'Ativo1': [100, 15, 99, 101, 103, 98],
    'Ativo2': [50, 51, 52, 49, 53, 47],
    'Ativo3': [200, 202, 198, 201, 205, 195]
})
```

```python
# Calcula o Z-score para cada ativo
df_zscore = df.apply(zscore)
```

```python
# Define um limiar para considerar um valor como outlier
limiar = 2
```

```python
# Identifica onde o valor absoluto do Z-score Ã© maior que o limiar
outliers = df_zscore > limiar
```

```python
# Cria um grÃ¡fico de dispersÃ£o para cada ativo
for coluna in df.columns:
    plt.scatter(df.index, df[coluna], label=coluna)
    # Destaca os outliers
    plt.scatter(df.index[outliers[coluna]], df[coluna][outliers[coluna]], color='red', label=f'Outliers {coluna}')
```
